```python
# ==========================================
# TASK 5 - Interactive Dashboard Development
# ==========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv(r"C:\Users\kumar\Downloads\ai_student_impact_dataset (1).csv")


# Display Dataset Information
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nFirst 5 Rows")
display(df.head())

# ------------------------------------------
# Data Cleaning
# ------------------------------------------

# Remove Duplicate Values
df = df.drop_duplicates()

# Fill Missing Values
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].mean())

print("\nMissing Values")
print(df.isnull().sum())

```

```python
# ------------------------------------------
# Dashboard Charts
# ------------------------------------------

# Numeric Columns
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

# Categorical Columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# ------------------------------------------
# Dashboard Charts
# ------------------------------------------

# Numeric Columns
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

# Categorical Columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# ===========================
# Chart 1 - Bar Chart
# ===========================

if len(categorical_cols) > 0:

    plt.figure(figsize=(8,5))

    df[categorical_cols[0]].value_counts().plot(kind='bar')

    plt.title("Category Distribution")
    plt.xlabel(categorical_cols[0])
    plt.ylabel("Count")

    plt.tight_layout()
    plt.show()

# ===========================
# Chart 2 - Pie Chart
# ===========================

if len(categorical_cols) > 0:

    plt.figure(figsize=(6,6))

    df[categorical_cols[0]].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')

    plt.ylabel("")
    
    plt.title("Top 5 Categories")

    plt.tight_layout()

    plt.show()
```

```python
# ===========================
# Chart 2 - Pie Chart
# ===========================

if len(categorical_cols) > 0:

    plt.figure(figsize=(6,6))

    df[categorical_cols[0]].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')

    plt.ylabel("")
    
    plt.title("Top 5 Categories")

    plt.tight_layout()

    plt.show()
```

```python
# ===========================
# Chart 3 - Histogram
# ===========================

if len(numeric_cols) > 0:

    plt.figure(figsize=(8,5))

    plt.hist(df[numeric_cols[0]],bins=20)

    plt.title(f"Distribution of {numeric_cols[0]}")

    plt.xlabel(numeric_cols[0])

    plt.ylabel("Frequency")

    plt.tight_layout()

    plt.show()
```

```python
# ===========================
# Chart 4 - Line Chart
# ===========================

if len(numeric_cols) > 0:

    plt.figure(figsize=(8,5))

    plt.plot(df[numeric_cols[0]])

    plt.title(f"{numeric_cols[0]} Trend")

    plt.xlabel("Index")

    plt.ylabel(numeric_cols[0])

    plt.tight_layout()

    plt.show()
```

```python

# ===========================
# Chart 5 - Correlation Heatmap
# ===========================

if len(numeric_cols) > 1:

    corr = df[numeric_cols].corr()

    plt.figure(figsize=(8,6))

    plt.imshow(corr)

    plt.colorbar()

    plt.xticks(range(len(corr.columns)),corr.columns,rotation=90)

    plt.yticks(range(len(corr.columns)),corr.columns)

    plt.title("Correlation Matrix")

    plt.tight_layout()

    plt.show()
```
